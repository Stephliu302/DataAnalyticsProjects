### Data Cleaning and Prepping

Recall that our goal is to predict whether or not a person makes over $50,000. We'll make a few adjustments to the data in order to prepare it for modeling.

Just to make the income categories a little clearer, we'll replace "<=50K" with "less than 50K" and ">50K" with "greater than 50K".

```python
income_data.loc[(income_data.income == ' <=50K'),'income']='less than 50K'
income_data.loc[(income_data.income == ' >50K'),'income']='greater than 50K'

# income category breakdown
print(income_data['income'].value_counts())
```

Let's also change the "?" classes in workclass, occupation and native-country to "unknown".

```python
income_data.loc[(income_data.workclass == ' ?'),'workclass']=' Unknown'
print(income_data['workclass'].value_counts())
print('---')

income_data.loc[(income_data.occupation == ' ?'),'occupation']=' Unknown'
print(income_data['occupation'].value_counts())
print('---')

income_data.loc[(income_data['native-country'] == ' ?'),'native-country']=' Unknown'
print(income_data['native-country'].value_counts())
print('---')
```

For the Neural Networks Modeling, we need all numeric inputs. Therefore, we should create dummy variables for the categorical variables. Since we will use the numeric coding for education, we will not create dummies for the categorical education variable.

```python
# create dummy variables for the categorical variables
income_data_clean = income_data.merge(pd.get_dummies(income_data[['workclass', 
                                                                  'marital-status',
                                                                  'occupation', 
                                                                  'relationship',
                                                                  'race',
                                                                  'sex', 
                                                                  'native-country']]), 
                                      left_index = True, right_index = True)

# shift column 'income' to first position
first_column = income_data_clean.pop('income') 
# insert column using insert(position,column_name,first_column) function
income_data_clean.insert(0, 'income', first_column)

income_data_clean.head()
```

We see that there are now 101 columns (variables) after each categorical variable has been changed into dummy variables.

Now, we would like to remove the original categorical variables, since we will not be including them in our models.

```python
income_data_clean = income_data_clean.drop(['workclass', 
                                            'education',
                                            'marital-status',
                                            'occupation', 
                                            'relationship',
                                            'race',
                                            'sex', 
                                            'native-country'], axis = 1)

income_data_clean.head()
```

After removing the catgeorical variables, we see there are now 93 variables (92 predictors and 1 target). 

In the neural network models, the input level will have 92 nodes. In addition, since income has 2 classes, there will be 2 nodes in the output level. The hidden levels and number of nodes in each will vary depending on the model we test.

## Modeling

Now we will try to find the best model that predicts/classifies income. We'll start by creating some neural networks and then compare their performance with some other modeling techniques.

```python
from numpy import random
random.seed(5)

# define datasets for modeling
X = income_data_clean.drop(['income'],axis=1)
Y = income_data_clean['income']

# split data into test (20%) and train sets (80%)
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, Y, test_size = 0.20, random_state = 5)
```

### Neural Networks

Let's set up a grid of potential values for some of the neural network parameters. We'll then fit neural network models for all the combinations of these parameter sets.

Some of the parameters we aim to adjust are:
- hidden_layer_sizes: number of hidden layers and nodes in each
- activation: activation function used for hidden layer
- solver: the solver for weight optimization
- alpha: L2 penalty parameter
- learning_rate: learning rate schedule for weight updates (in this lab, "constant" and "adaptive" typically result in similar accuracy rates)


```python
from sklearn.neural_network import MLPClassifier
random.seed(5)

# define MLP Classifier
mod1 = MLPClassifier(max_iter=100, random_state=5)

# create grid of parameter options
parameter_space = {
    'hidden_layer_sizes': [(100,), (100, 100), (100, 100, 100)],
    'activation': ['relu', 'tanh'],
    'solver': ['adam', 'lbfgs'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant'],
}
```

```python
from pandas.core.common import random_state
from sklearn.utils.extmath import randomized_svd
from sklearn.model_selection import GridSearchCV
random.seed(5)

# perform grid search using all the parameters above (with 3 splits)
clf = GridSearchCV(mod1, parameter_space, n_jobs = -1, cv = 3)
clf.fit(train_X, train_y)
```

Now that we've fit all the models to the train set, we can get the results to determine which set of parameters performed the best. We'll look at the accuracy for each set of parameters.

```python
random.seed(5)
# All results
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))
```

```python
# get best performing results 
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
  if mean > 0.77:
    print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))
```

```python
random.seed(5)
# Show best parameter set
print('Best parameters found:\n', clf.best_params_)
```

We can see that the best Neural Network model uses:
- Activation Function: relu (rectified linear unit function)
- alpha = 0.05
- 3 hidden layers with 100 nodes in each
- contant learning rate
- adam solver (stochastic gradient-based optimizer)

It has an **accuracy of 0.799**.

Now that we've found the best performing neural network, let's use the model to predict on the test set.

```python
from sklearn.metrics import accuracy_score
random.seed(5)

# make predictions
pred_y = clf.predict(test_X)

NN_accuracy = accuracy_score(test_y, pred_y)
print('Accuracy: %.2f%%' % (NN_accuracy * 100))

from sklearn.metrics import classification_report
print('Results on the test set:')
print(classification_report(test_y, pred_y))
```

From the best model with the parameter set discussed above, we see an overall **80% accuracy**. This is somewhat high, so we could use it to predict whether an individual's income is above $50,000. However, we can also explore other types of models that may predict with higher accuracy.

### Boosting

Let's compare the performance of the neural networks above to some boosting models.

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
random.seed(5)

# AdaBoost Classifier
AdaBoost = AdaBoostClassifier(n_estimators=500, random_state=5)
# fit model to train data
AdaBoost.fit(train_X, train_y)

# predict model on test data
Ada_preds = AdaBoost.predict(test_X)

# evaluate model
Ada_accuracy = accuracy_score(test_y, Ada_preds)
print('The accuracy is: ',Ada_accuracy*100,'%')
```

We can see that the AdaBoost classifier predicts with **87% accuracy**. This is better than the best performing neural networks model.

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
random.seed(5)

# define XGBoost classifier
XGB_model = XGBClassifier(learning_rate=0.5, min_child_weight=6, random_state=5)
XGB_model.fit(train_X, train_y)

# predict model on test data
XGB_preds = XGB_model.predict(test_X)

# evaluate model
XGB_accuracy = accuracy_score(test_y, XGB_preds)
print('Accuracy: %.2f%%' % (XGB_accuracy * 100))
```

The XGBoost classifier performs about the same as the AdaBoost classifier, with only slightly higher accuracy (**87.4%**). Both Boosting Models predict with much higher accuracy than the Neural Network Models. It would probably be better to use one of these boosting models for predicting income.

Note that boosting models are iterative, which means they continually try to improve themselves until they get the best possible model. Because of this, it is likely the boosting models would out-perform the neural networks.

### Bagging

Let's do the same as we did above with the boosting models, but now using a bagging ensemble model to predict income.

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
random.seed(5)

# Bagging Classifier
Bag = BaggingClassifier(n_estimators = 100, random_state=5)
# fit model to train data
Bag.fit(train_X, train_y)

# predict model on test data
Bag_preds = Bag.predict(test_X)

# evaluate model
Bag_accuracy = accuracy_score(test_y, Bag_preds)
print('The accuracy is: ',Bag_accuracy*100,'%')
```

The Bagging Classifier also performs better than the best Neural Network model, with an **accuracy of 85%**. However, it still does not perform as well as the boosting models.
